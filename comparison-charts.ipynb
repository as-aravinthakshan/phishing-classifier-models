```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
csv_path = 'model_performance.csv'  # Replace with the path to your CSV file
df = pd.read_csv(csv_path)

# Plot the line graph
plt.figure(figsize=(11, 6))
plt.plot(df['Model'], df['F1 Train Score'], marker='o', label='F1 Score Train',color ='blue')
plt.plot(df['Model'], df['F1 Test Score'], marker='x', label='F1 Score Test',color ='red')

# Adding title and labels
plt.title('Model Performance Comparison')
plt.xlabel('Model')
plt.ylabel('F1 Score')
plt.xticks(rotation=90)
plt.legend()

# Annotate each point with the F1 score
for i, txt in enumerate(df['F1 Train Score']):
    plt.annotate(f'{txt:.4f}', (df['Model'][i], df['F1 Train Score'][i]), textcoords="offset points", xytext=(0,5), ha='center')
for i, txt in enumerate(df['F1 Test Score']):
    plt.annotate(f'{txt:.4f}', (df['Model'][i], df['F1 Test Score'][i]), textcoords="offset points", xytext=(0,-15), ha='center')

# Display the plot
plt.show()
```
